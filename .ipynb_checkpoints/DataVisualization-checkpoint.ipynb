```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/Case3HistoricalPrices.csv")
data.head()
```

## Distribution of values

```python
fig, axes = plt.subplots(nrows=6, ncols=4, figsize=(30,30))
for i, ax in enumerate(axes.flatten()):
    name = data.columns[i+1]
    y = data[name]
    ax.hist(y, bins=50)
    ax.title.set_text(name)
    ax.set_xlabel("Value")
    ax.set_ylabel("Occurences")
```

## Line Graph

```python
x = data[data.columns[0]].to_numpy()
fig, axes = plt.subplots(nrows=8, ncols=3, figsize=(30,60), sharex=True, sharey=True)
for i, ax in enumerate(axes.flatten()):
    name = data.columns[i+1]
    y = data[name]
    ax.plot(x, y)
    ax.set_title(name, fontweight="bold", size=20)
    ax.set_xlabel("Time", size=20)
    ax.set_ylabel("Price", size=20)
```

## Simple moving average

```python
x = data[data.columns[0]].to_numpy()
fig, axes = plt.subplots(nrows=8, ncols=3, figsize=(30,60), sharex=True, sharey=True)
for i, ax in enumerate(axes.flatten()):
    name = data.columns[i+1]
    y = data[name]
    line1, = ax.plot(x, y, 'c')
    y = data.iloc[:,i+1].rolling(window=252).mean()
    line2, = ax.plot(x, y, 'b')
    y = data.iloc[:,i+1].rolling(window=21).mean()
    line3, = ax.plot(x, y, 'r')
    ax.legend((line1, line2, line3), ('Original', 'Yearly', 'Monthly'))
    ax.set_title(name,fontweight="bold", size=20)
    ax.set_xlabel("Time", size=20)
    ax.set_ylabel("Price", size=20)
```

## Absolute change per day

```python
x = data[data.columns[0]].to_numpy()
fig, axes = plt.subplots(nrows=8, ncols=3, figsize=(30,60), sharex=True, sharey=True)
for i, ax in enumerate(axes.flatten()):
    name = data.columns[i+1]
    y = data.iloc[:,i+1].diff()
    ax.plot(x, y)
    ax.set_title(name,fontweight="bold", size=20)
    ax.set_xlabel("Time", size=20)
    ax.set_ylabel("Absolute Change", size=20)
```

## Percent change per day

```python
x = data[data.columns[0]].to_numpy()
fig, axes = plt.subplots(nrows=8, ncols=3, figsize=(30,60), sharex=True, sharey=True)
for i, ax in enumerate(axes.flatten()):
    name = data.columns[i+1]
    y = data.iloc[:,i+1].pct_change()
    ax.plot(x, y)
    ax.set_title(name,fontweight="bold", size=20)
    ax.set_xlabel("Time", size=20)
    ax.set_ylabel("Percentage Change", size=20)
```

## Lag Plot (w/ Lag 1)

```python
fig, axes = plt.subplots(nrows=8, ncols=3, figsize=(30,60))
for i, ax in enumerate(axes.flatten()):
    pd.plotting.lag_plot(data.iloc[:, i+1], lag=1, ax=ax)
    ax.set_title(name,fontweight="bold", size=20)
    ax.set_xlabel("Price(t)", size=20)
    ax.set_ylabel("Price(t+1)", size=20)
```

## Autocorrelation Plot

```python
fig, axes = plt.subplots(nrows=8, ncols=3, figsize=(30,60))
for i, ax in enumerate(axes.flatten()):
    pd.plotting.autocorrelation_plot(data.iloc[:, i+1], ax=ax)
    ax.set_title(name,fontweight="bold", size=20)
    ax.set_xlabel("Lag", size=20)
    ax.set_ylabel("Autocorrelation", size=20)
```

## Frequency Domain

Lol idk what this shows I just thought it seemed cool -- seems like there's no seasonality in the data though

```python
import scipy.fftpack as fftpack

fig, axes = plt.subplots(nrows=8, ncols=3, figsize=(30,60), sharex=True)
for i, ax in enumerate(axes.flatten()):
    name = data.columns[i+1]
    y = data[name].to_numpy()
    y_fft = fftpack.fft(y)
    x = fftpack.fftfreq(len(x))
    ax.plot(x, np.abs(y_fft))
    ax.set_xlim(-0.1, 0.1)
    ax.set_title(name, fontweight="bold", size=20)
    ax.set_xlabel("Frequency", size=20)
    ax.set_ylabel("Magnitude", size=20)
```
